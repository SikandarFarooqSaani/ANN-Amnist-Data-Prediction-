```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape #60000 images, 28*28
```

```python
x_train #pixel value of images
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[2]) #image
```

```python
#neural network
x_train[13]
```

```python
#we have to scale values from 0 to 1 so their values are in scale so weights can be easily calculated
#devide each with 255

x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
model = Sequential()

#problem first layer will have 784 layer but data is in 28 *28 so we will have to flatten it
#we have imported flatten

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu')) #no need to tell input
model.add(Dense(32,activation='relu'))
#model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
#no need to ohe sparse_categorical_crossentropy
```

```python
#model compiled
history = model.fit(x_train,y_train,epochs=15,validation_split=0.2)
```

```python
#prediction will calculate probility of every iteration
y_prob = model.predict(x_test)
```

```python
y_pred = y_prob.argmax(axis=1) #test image result
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_accuracy'])
```
